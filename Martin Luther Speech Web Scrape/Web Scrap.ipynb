# Web Scraper + Regular Expression + Pandas

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'http://analytictech.com/mb021/mlk.htm'
requests.get(url)
```

```python
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
speech = soup.find_all('p')
speech
```

```python
type(speech)
```

```python
speech_combined = [p.text for p in speech]
speech_combined
```

```python
type(speech_combined)
```

```python
string_speech = ' '.join(speech_combined)
string_speech
```

```python
type(string_speech)
```

```python
string_speech[:100]
```

```python
string_speech_cleaned = string_speech.replace('\r\n', ' ')
string_speech_cleaned
```

```python
import re
```

```python
puntuations = re.sub(r'[\w\s]', '', string_speech_cleaned)
puntuations
```

```python
speech_no_puntuations = re.sub(r'[^\w\s]', '', string_speech_cleaned)
print(speech_no_puntuations)
```

```python
speech_lower = speech_no_puntuations.lower()
print(speech_lower)
```

```python
re.split(r'\s+', speech_lower)
```

```python
spech_words_broken_out = re.split(r'\s+', speech_lower)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(spech_words_broken_out).value_counts()
df
```
