# News Detection

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import re
import string
```

```python
fake_news = pd.read_csv("fake news.csv", header=0, sep=",")
true_news = pd.read_csv("True news.csv", header=0, sep=",")
```

```python
#head shows first five while tail shows last five
fake_news.head()
```

```python
fake_news["class"]=0
true_news["class"]=1
```

```python
fake_news.shape, true_news.shape
```

```python
#removing the last 10 rows
fake_news_testing = fake_news.tail(10)
for i in range(23480, 23470, -1):
    fake_news.drop([i], axis=0, inplace=True)

true_news_testing = true_news.tail(10)
for i in range(21416, 21406, -1):
    true_news.drop([i], axis=0, inplace=True)
```

```python
fake_news.shape, true_news.shape
```

```python
fake_news_testing['class'] = 0 
true_news_testing['class'] = 1 
```

```python
fake_news_testing.head(10)
```

```python
true_news_testing.head(10)
```

```python
#to merge data
news_merge = pd.concat([fake_news, true_news], axis = 0)
news_merge.head(10)
```

```python
#putting the removed 10 values in csv - more like exporting it
news_merge.to_csv("merged_news.csv")
```

```python
news_merge.shape
```

```python
news_merge.columns
```

```python
# I removed columns I do not need
data = news_merge.drop(['title','subject','date'], axis = 1)
```

```python
print(data)
```

```python
data.isnull().sum()
```

```python
data = data.sample(frac = 1)
data.head()
```

```python
data.reset_index(inplace = True)
data.drop(['index'], axis = 1, inplace = True)
```

```python
data.columns
```

```python
data.head()
```

```python
def wordopt(text):
    text = re.sub(r'\[.*?]', '', text)
    text = re.sub(r'\W', ' ', text)
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'<.*?>', '', text)
    text = re.sub(r'[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub(r'\n', '', text)
    text = re.sub(r'\w*\d\w*', '', text)
    return text
```

```python
data['text'] = data['text'].apply(wordopt)
```

```python
x = data['text']
y = data['class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train) 
xv_test = vectorization.transform(x_test)
```

### Logistic Regresssion

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(xv_train, y_train)
```

```python
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_lr))
```

##Decision Tree classification

```python
from sklearn.tree import DecisionTreeClassifier
DT = DecisionTreeClassifier()
DT.fit(xv_train, y_train)
```

```python
pred_dt = DT.predict(xv_test)
```

```python
DT.score(xv_test,y_test)
```

```python
print(classification_report(y_test, pred_dt))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
GB = GradientBoostingClassifier(random_state = 0)
GB.fit(xv_train, y_train)
```

```python
pred_gb = GB.predict(xv_test)
```

```python
GB.score(xv_test,y_test)
```

```python
print(classification_report(y_test, pred_gb))
```

```python
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier(random_state = 0)
RF.fit(xv_train, y_train)
```

```python
pred_rf = RF.predict(xv_test)
```

```python
RF.score(xv_test,y_test)
```

```python
print(classification_report(y_test, pred_rf))
```

###Manual Testing

```python
def output_lable(n):
    if n == 0:
        return "Fake News"
    elif n == 1:
        return "Not A Fake News"
    
def manual_testing(news):
    testing_news = {"text":[news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt) 
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)
    pred_LR = LR.predict(new_xv_test)
    pred_DT = DT.predict(new_xv_test)
    pred_GB = GB.predict(new_xv_test)
    pred_RF = RF.predict(new_xv_test)

    return print("\n\nLR Prediction: {} \nDT Prediction: {} \nGB Prediction: {} \nRF Prediction: {}".format(output_lable(pred_LR[0]), 
                                                                                                              output_lable(pred_DT[0]), 
                                                                                                              output_lable(pred_GB[0]), 
                                                                                                              output_lable(pred_RF[0])))
```

```python
news = str(input())
manual_testing(news)
```
