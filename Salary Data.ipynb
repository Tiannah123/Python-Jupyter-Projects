# Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
salary_data = pd.read_csv("Salary_Data.csv")
x = salary_data.iloc[:, :-1]. values
y = salary_data.iloc[:, 1].values
```

```python
#splitting the data set into test and train. 1/3 for test purpose and 2/3 for train purpose
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 1/2, random_state = 0)
```

```python
#linear regresstion for train section; common in all algorithms
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
#predit the values; common in all algorithms
y_pred = lr.predict(x_test)
y_pred
```

```python
plt.scatter(x_train, y_train, color = 'blue')
plt.plot(x_train, lr.predict(x_train), color = 'red')
plt.title('salary-Experience(Train set)')
plt.xlabel('Years of Experiance')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(x_test, y_test, color = 'blue')
plt.plot(x_test, lr.predict(x_test), color = 'red')
plt.title('salary-Experience(Test set)')
plt.xlabel('Years of Experiance')
plt.ylabel('Salary')
plt.show()
```

```python
#calculating residuals
from sklearn import metrics 
print ('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print ('MSE:', metrics.mean_squared_error(y_test, y_pred))
print ('RMSE:', np.sqrt(metrics.mean_absolute_error(y_test, y_pred)))
```
